```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def corr(s, D):
    varx = D/3
    return 1/np.sqrt(1+2*(s**2/varx) + (s**2/varx)**2)
```

```python
corr(0.025, 1)
```

```python
corr(0.025, 0.25)
```

```python
plt.figure()

sigmas = []
for s in np.linspace(0, 0.6, 8):
    ds = np.linspace(0.00001, 1, 100)
    ps = [corr(s, D) for D in ds]
    sigmas.append(s)
    plt.plot(ds, ps)
plt.show()
```
